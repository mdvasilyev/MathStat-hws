```python
# import libs
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define true distribution
def true_distribution(x):
    return 1 - np.exp(-x)
```

```python
# define empirical distribution
def empirical_distribution(sample, x):
    n = len(sample)
    count = np.zeros_like(x, dtype=float)
    
    for value in sample:
        count += x >= value
    
    return count / n
```

```python
def simulate(num_samples, sample_size):
    # Plot true distribution
    x = np.linspace(0, 15, 1000)
    plt.plot(x, true_distribution(x), label='True Distribution', color='blue')

    # Sample from exponential distribution and plot empirical distributions
    for _ in range(num_samples):
        sample = np.random.exponential(size=sample_size)
        edf = empirical_distribution(sample, x)
        plt.plot(x, edf, alpha=0.2, color='red')

    plt.plot(x, edf, alpha=0.2, color='red', label='Empirical Distributions')
    plt.title('Exponential Distribution')
    plt.ylabel('Cumulative Probability')
    plt.legend()
    plt.show()
```

```python
simulate(num_samples=50, sample_size=50)
```
